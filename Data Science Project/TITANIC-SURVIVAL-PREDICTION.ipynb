```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
import io
titanic = pd.read_csv(io.BytesIO(uploaded['tested (1).csv']))
```

```python
column_train=['Age','Pclass','SibSp','Parch','Fare','Sex','Embarked']
x=titanic[column_train]
y=titanic['Survived']
```

Step no.4 - Train and Test Split

In order to evaluate the performance of our model, we need to split the dataset into training and testing sets. We will use 80% of the data for training and 20% for testing.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Step no.5 - Build and Train the Model

Now, we can build our machine learning model using the linear regression from scikit-learn and train it on the training data.

```python
model = HistGradientBoostingClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

This code uses a Hist Gradient Boosting Classifier to train the model and predicts the survival status of the passengers. The accuracy score is then printed to evaluate the model's performance.
